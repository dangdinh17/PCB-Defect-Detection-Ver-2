```python
import os
from PIL import Image
# Đường dẫn đến thư mục chứa các tệp
source_folder = 'real_test/test1_600x600\images'
destination_folder = 'real_test/test1_150x150/images'

# Các từ khóa cần đếm
for filename in os.listdir(source_folder):
    if filename.endswith(".jpg") or filename.endswith(".png"):  # Chỉ xử lý các tệp ảnh
        # Đường dẫn đầy đủ của tệp gốc
        source_file = os.path.join(source_folder, filename)
        # Đường dẫn đầy đủ của tệp đích
        destination_file = os.path.join(destination_folder, filename[:-7] + '150.jpg')
        
        # Mở ảnh và thay đổi kích thước
        with Image.open(source_file) as img:
            img_resized = img.resize((150, 150), Image.Resampling.LANCZOS)
            # Lưu ảnh vào thư mục đích
            img_resized.save(destination_file)
# In kết quả

```

```python
import os
import shutil

# Đường dẫn đến thư mục chứa các tệp gốc và thư mục đích
source_folder = 'real_test/test1_150x150\labels'
destination_folder = 'real_test/test1_EDSRx4/labels'

# Tạo thư mục đích nếu chưa tồn tại
os.makedirs(destination_folder, exist_ok=True)

# Duyệt qua các tệp trong thư mục gốc
for filename in os.listdir(source_folder):
    if filename.endswith(".txt"):  # Chỉ xử lý các tệp .txt
        # Đường dẫn đầy đủ của tệp gốc
        source_file = os.path.join(source_folder, filename)
        # Đường dẫn đầy đủ của tệp đích
        destination_file = os.path.join(destination_folder, filename[:-4] + '_edsrx4.txt')
        
        # Sao chép tệp vào thư mục đích
        shutil.copy2(source_file, destination_file)

print("Hoàn thành việc sao chép các tệp .txt.")
```

```python
import os

# Đường dẫn đến các thư mục cần liệt kê
image_folder = 'real_test/test1_EDSRx2/images'
label_folder = 'real_test/test1_EDSRx2/labels'

# Lấy danh sách các tệp trong các thư mục
image_files = os.listdir(image_folder)
label_files = os.listdir(label_folder)

# Sắp xếp danh sách các tệp
image_files.sort()
label_files.sort()

# Duyệt và in từng tệp cùng lúc từ cả hai danh sách
print("Tệp trong thư mục 'images' và 'labels':")
for image_file, label_file in zip(image_files, label_files):
    print(f"Image file: {image_file} | Label file: {label_file}")
```
